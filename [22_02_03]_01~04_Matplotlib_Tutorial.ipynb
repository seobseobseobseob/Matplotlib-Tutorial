# 01. Matplotlib 기본 사용
---
* Keyword: matplotlib.pyplot, plt.plot(), plt.show(), plt.axis(), 포맷 문자열

## 기본 그래프 그리기

image.png

```python
import matplotlib.pyplot as plt

# 리스트가 하나만 주어지면, 그것들을 y값이라고 가정하고 그래프를 그려줌
plt.plot([1,2,3,4])
plt.show()
```

```python
# 리스트가 두개 주어질 경우, 앞이 x이고 뒤가 y이다.
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

## 스타일 지정하기

image.png

```python
# plot의 세번째 인자에 포맷 문자열로 선의 색상과 형태를 지정
plt.plot([1,2,3,4], [1,3,9,15],'ro')
# axis는 축의 범위를 지정
# 리스트에서 앞의 두 인자는 x축, 뒤의 두 인자는 y축
plt.axis([0,6,0,20])
plt.show()
```

```python
plt.plot([1,2,3,4], [1,3,9,15],'r')
plt.axis([0,6,0,20])
plt.show()
```

```python
plt.plot([1,2,3,4], [1,3,9,15],'r--')
plt.axis([0,6,0,20])
plt.show()
```

##  여러 개의 그래프 그리기

image.png

```python
import numpy as np
t = np.arange(0.,5.,0.2)

plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^', t,t**4,'--')
plt.show()
```

# 02. Matplotlib 숫자 입력하기
---
* Keyword: matplotlib.pyplot, plt.plot()

## 레이블이 있는 데이터 사용하기
image.png

```python
data = {'data_x' : [x for x in range(6)], 'data_y': [2,3,5,10,8, 7]}
```

```python
# 데이터 레이블을 입력하여 그래프를 그릴 수 있음
plt.plot('data_x','data_y',data = data)
plt.show()
```

# 03. Matplotlib 축 레이블 설정하기
---
* Keyword: plt.xlabel(), plt.ylabel(), plt.axis(), labelpad, fontdict, loc

## 기본 사용
image.png

```python
plt.plot([1,2,3,4],[2,4,8,10],'--x')
plt.xlabel('X_Label')
plt.ylabel('Y,Label')
plt.show()
```

## 여백 지정하기
image.png

```python
plt.plot([1,2,3,4],[2,4,6,10])
plt.xlabel('X-Axis',labelpad=15)
plt.ylabel('Y-Axis', labelpad=20)
plt.show()
```

## 폰트 설정하기
image.png

```python
plt.plot([1,2,3,4],[2,3,5,10])
plt.xlabel('X-Axis', labelpad = 15, fontdict = {'family' : 'serif', 'color':'b','weight':'bold','size':14})
plt.ylabel('Y-Axis', labelpad = 20, fontdict = {'family' : 'fantasy', 'color':'deeppink','weight':'normal','size':'xx-large'})
plt.show()
```

```python
plt.plot([1,4,6,9],[1,5,7,2])

font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }

plt.xlabel('X', fontdict = font1)
plt.ylabel('Y', fontdict = font2)
plt.show()
```

# 04. Matplotlib 범례 표시하기
Keyworkd: plt.legend(), label, loc, ncol, fontsize, frameon, shadow, 범례

## 기본 사용
image.png

```python
plt.plot([1,2,3,4],[2,3,5,10],label = 'Price')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## 위치 지정하기
image.png

```python
plt.plot([1,2,3,4],[2,3,5,10],label = 'Price')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = (0.5,0.5))
plt.show()
```

```python
plt.plot([1,2,3,4],[2,3,5,10],label = 'Price')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left')
plt.show()
```

```python
plt.plot([1,2,3,4],[2,3,5,10],label = 'Price')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = (0,0))
plt.show()
```

```python
plt.plot([1,2,3,4],[2,3,5,10],label = 'Price')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=(1.0,1.0))
plt.show()
```

## 열 개수 지정하기
image.png

```python
plt.plot([1,2,3,4],[2,3,5,10],label = 'Price1')
plt.plot([1,2,3,4],[4,14,2,16], label = 'Price2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.plot([1,2,3,4],[2,3,5,10],label = 'Price1')
plt.plot([1,2,3,4],[4,14,2,16], label = 'Price2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(ncol = 2)
plt.show()
```

## 폰트 크기 지정하기
image.png

```python
plt.plot([1,2,3,4],[2,3,5,10],label = 'Price1')
plt.plot([1,2,3,4],[4,14,2,16], label = 'Price2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(ncol = 2, fontsize = 15)
plt.show()
```

## 범례 테두리 꾸미기
image.png

```python
plt.plot([1,2,3,4],[2,3,5,10],label = 'Price1')
plt.plot([1,2,3,4],[4,14,2,16], label = 'Price2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(ncol = 2, fontsize = 15, frameon=True, shadow = True)
plt.show()
```

```python
plt.plot([1,2,3,4],[2,3,5,10],label = 'Price1')
plt.plot([1,2,3,4],[4,14,2,16], label = 'Price2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(ncol = 2, fontsize = 15, frameon=False, shadow = True)
plt.show()
```
